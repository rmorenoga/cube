## Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Import csv to pandas table

```python
readings = pd.read_csv("WeightTest.csv",index_col=0)
display(readings)
```

## General Test Graph

```python
ax = sns.lineplot(data=readings["Pressure"],markers=True)
ax.set_xlabel("Reading")
ax.set_ylabel("Pressure (hPa)")
```

## Grouped by weight

```python
f, ((ax1,ax2,ax3)) = plt.subplots(1, 3,figsize=(10, 3))

grouped = readings.groupby(['Position'])
df0 = grouped.get_group(0)
sns.lineplot(data=df0,x="Weight",y ="Pressure",markers=True,ax = ax1)
ax1.set_xlabel("Weight (g)")
ax1.set_ylabel("Pressure (hPa)")
ax1.set_ylim([1005,1033])
ax1.set_title("Position 0 mm off top")

df5 = grouped.get_group(5)
sns.lineplot(data=df5,x="Weight",y ="Pressure",markers=True,ax = ax2)
ax2.set_xlabel("Weight (g)")
ax2.get_yaxis().set_visible(False)
ax2.set_ylim([1005,1033])
ax2.set_title("Position 5 mm off top")

df10 = grouped.get_group(10)
sns.lineplot(data=df10,x="Weight",y ="Pressure",markers=True,ax = ax3)
ax3.set_xlabel("Weight (g)")
ax3.get_yaxis().set_visible(False)
ax3.set_ylim([1005,1033])
ax3.set_title("Position 10 mm off top")

#f15 = grouped.get_group(15)
#sns.lineplot(data=df15,x="Weight",y ="Pressure",markers=True,ax = ax4)
#ax4.set_xlabel("Reading")
#ax4.set_ylim([1005,1033])
#ax4.set_title("Position 15 mm off top")
```
